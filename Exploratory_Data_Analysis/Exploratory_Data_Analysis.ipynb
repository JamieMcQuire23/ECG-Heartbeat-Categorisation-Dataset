```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Exploratory Data Analysis

* This notebook contains the exploratory data analysis for the ECG Heartbeat Catergorization Dataset.
* The dataset contains two datasets of pre-processed heartbeat signals taken from the MIT-BIH Arrhythmia Dataset and The PTB Diagnostic ECG Database.
* The goal of the project is to identify myocardial infraction (heart attack).

# MIT-BIH Arrhythmia Dataset

```
#read in the mit dataset
mit_train = pd.read_csv("/content/drive/My Drive/ECG_HeartBeat_Pytorch/Data/heartbeat/mitbih_train.csv",header=None)
mit_test = pd.read_csv("/content/drive/My Drive/ECG_HeartBeat_Pytorch/Data/heartbeat/mitbih_test.csv",header=None)

#print the dim
print("Training Data: \n")
print("Number of columns: %d \nNumber of rows: %d" % (mit_train.shape[0],mit_train.shape[1]))

print("\nTesting Data: \n")
print("Number of columns %d \nNumber of rows: %d" % (mit_test.shape[0],mit_test.shape[1]))
```

# Visualisation

```
vals = mit_train.values
X_train = vals[:, :-1]
Y_train = vals[:, -1].astype(int)
```

* We can see that there is imbalanced classes which we will have to account for later when training the deep learning models.
* The code from this section was taken from https://www.kaggle.com/coni57/model-from-arxiv-1805-00794

```
C0 = np.argwhere(Y_train==0).flatten()
C1 = np.argwhere(Y_train==1).flatten()
C2 = np.argwhere(Y_train==2).flatten()
C3 = np.argwhere(Y_train==3).flatten()
C4 = np.argwhere(Y_train==4).flatten()


print(C0.shape)
print(C1.shape)
print(C2.shape)
print(C3.shape)
print(C4.shape)
```

```
x = np.arange(0, 187)*8/1000

plt.figure(figsize=(10,6))
plt.plot(x, X_train[C0, :][0], label="Cat. N")
plt.plot(x, X_train[C1, :][0], label="Cat. S")
plt.plot(x, X_train[C2, :][0], label="Cat. V")
plt.plot(x, X_train[C3, :][0], label="Cat. F")
plt.plot(x, X_train[C4, :][0], label="Cat. Q")
plt.legend()
plt.title("1-beat ECG for every category", fontsize=20)
plt.ylabel("Amplitude", fontsize=15)
plt.xlabel("Time (ms)", fontsize=15)
plt.show()
```
